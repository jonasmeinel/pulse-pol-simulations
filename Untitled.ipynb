```python
import numpy as np
import matplotlib.pyplot as plt
import qutip as q 
```

```python
Azz=2.2 * 2 * np.pi #MHz
Azx=0.2 * 2 * np.pi #MHz
B0=1.0 #mT
bi = 10 * 2 * np.pi #MHz

Omega = 2 * np.pi * 10

time_x= 1
npoints = 500
t = np.linspace(0, time_x, npoints)

#NV operators
#electron reduced as spin 1/2
sz = q.tensor(q.sigmaz()*0.5,q.qeye(2))
sx = q.tensor(q.sigmax()*0.5,q.qeye(2))
sy = q.tensor(q.sigmay()*0.5,q.qeye(2))

#nuclear as spin 1
iz = q.tensor(q.qeye(2),q.spin_Jz(0.5))
ix = q.tensor(q.qeye(2),q.spin_Jx(0.5))
iy = q.tensor(q.qeye(2),q.spin_Jy(0.5))

#initialize density matrix
phi = q.basis(2,0) # + q.basis(2,1))/np.sqrt(2)
rho_nv = phi*phi.dag()

#Nuclear initial pure state
rho_nuclear = q.thermal_dm(2, 1000)
rho_total = q.tensor(rho_nv,rho_nuclear)

#perfect flip-flop interaction
Delta = 0
H = (Delta + Omega) * sx + Azz * sz * ix + Omega * iz

resultat = []
resultat.append(q.mesolve(H=H, rho0=rho_total, tlist=t, c_ops = [], e_ops=[], options=q.Options(nsteps=100000)))    

plt.title('Flip-Flop, Hartmann Hahn')
plt.plot(t,q.expect(resultat[0].states, sx),label='$S_x$')
plt.plot(t,q.expect(resultat[0].states, iz),label='$I_z$')
plt.xlabel('time')
plt.ylabel('exp. value')
plt.legend()
```

We split the evolution in segments which are themselves independant: \
pi/2x - .... - pi/2x - pi/2y - .... - pi/2y - ...

```python
w_i = 2 * np.pi * 7 # 5.405 #MHz

Azz = 2 * np.pi * 0.4 # MHz
Azx = 2 * np.pi * 0.09 # MHz
H0 = Azz * iz * sz + w_i * iz + Azx * ix * sz 

def pulse(strength = 10.0, phase = 0):
    omega = strength * 2 * np.pi
    H = omega * (np.cos(np.deg2rad(phase)) * sx + np.sin(np.deg2rad(phase)) * sy ) + H0
    return H
 
def get_sequence(N,tevo):
    tpi = 0.05
    tpi2 = 0.025
    tevo_prime = tevo - tpi
    if tevo_prime < 0: return
    Hs = []
    ts = []
    for k in range(N):
        for i in range(2):
            Hs.append(pulse(phase=90*i))
            ts.append(tpi2)
            for j in range(4):
                Hs.append(H0)
                ts.append(tevo_prime*0.5)
                Hs.append(pulse(phase=90 * (j%2)))
                ts.append(tpi)
                Hs.append(H0)
                ts.append(tevo_prime*0.5)
            Hs.append(pulse(phase=90*(i)))
            ts.append(tpi2)
    return Hs, np.asarray(ts)

Hs,ts = get_sequence(30,0.185)

plt.title('sequence')
fids = np.where(np.abs(ts - 0.035)<1e-3 )
pis = np.where(np.abs(ts - 0.05)<1e-3 )
pi2s = np.where(np.abs(ts - 0.025)<1e-3 )

plt.plot(np.arange(len(ts))[fids],ts[fids],'o',label='fid')
plt.plot(np.arange(len(ts))[pis],ts[pis],'o',label='pi')
plt.plot(np.arange(len(ts))[pi2s],ts[pi2s],'o',label='pi2')
plt.plot(np.arange(len(ts)),ts,'--',color='grey')
plt.ylim([0.015,0.055])
plt.legend()
plt.show()

```

```python
def evolve(hs,ts,rho0):
    resultat = []
    for i, tau in enumerate(ts):
        try:
            npoints = 50
            t = np.linspace(0, tau, npoints)
            if i%1 ==0:
                resultat.append(q.mesolve(H=hs[i], rho0=resultat[-1].states[-1], tlist=t,
                                          c_ops = [], e_ops=[], options=q.Options(nsteps=100000)))
                #print('first index',i)
        except:
            npoints = 50
            t = np.linspace(0, tau, npoints)
            resultat.append(q.mesolve(H=hs[i], rho0=rho0, tlist=t,
                                      c_ops = [], e_ops=[], options=q.Options(nsteps=100000)))
    return resultat

#initialize density matrix
phi = (q.basis(2,0))# + q.basis(2,1))/np.sqrt(2)
rho_nv = phi*phi.dag()

#Nuclear initial pure state
rho_nuclear = q.thermal_dm(2, 1000)
phi = (q.basis(2,0)) # + q.basis(2,1))/np.sqrt(2)
rho_nuc = phi*phi.dag()

rho_total = q.tensor(rho_nv,rho_nuc)

Hs,ts = get_sequence(50,.2107)
resultats = evolve(Hs[:],ts[:],rho_total)
```

```python
t_offset = 0
for j,result in enumerate(resultats):
    npoints = 50
    t = np.linspace(0, ts[j], npoints)
    if j == 0:
        plt.plot(t,q.expect(resultats[j].states, iz),label='$S_x$')
        #plt.plot(t,q.expect(resultats[j].states, sz),label='$S_x$')
    else:
        plt.plot(t+t_offset,q.expect(resultats[j].states, iz),label='$S_x$')
        #plt.plot(t+t_offset,q.expect(resultats[j].states, sz),label='$S_x$')
    t_offset += t[-1]
```

```python
taus = np.linspace(0.055,0.25,250)
final_results = []
for k, tau in enumerate(taus):
    print(k)
    Hs,ts = get_sequence(10,tau)
    resultats = evolve(Hs[:],ts[:],rho_total)
    final_results.append(q.expect(resultats[-1].states, iz)[-1])
```
